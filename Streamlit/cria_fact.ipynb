### Criação da Query que filtrará o BD original e carregará o arquivo pronto para uso

```python
import duckdb

con = duckdb.connect()

query = """
COPY (
    SELECT
        PA_CODUNI,
        Ano,
        Mes,
        data_ref,
        PA_MUNPCN,
        PA_PROC_ID,

        SUM(CAST(PA_VALPRO AS DOUBLE)) AS PA_VALPRO,
        SUM(CAST(PA_VALAPR AS DOUBLE)) AS PA_VALAPR,
        SUM(CAST(PA_QTDPRO AS BIGINT)) AS PA_QTDPRO,
        SUM(CAST(PA_QTDAPR AS BIGINT)) AS PA_QTDAPR

    FROM read_parquet('C:/Projetos/PyPAH/dados_sia/dados_prontos/dados_trata.parquet')
    GROUP BY
        PA_CODUNI,
        Ano,
        Mes,
        data_ref,
        PA_MUNPCN,
        PA_PROC_ID
) TO '../dados_sia/facts_dash/fact_qtd_val.parquet'
(FORMAT PARQUET);
"""

con.execute(query)

print("FACT criada com sucesso")
```
